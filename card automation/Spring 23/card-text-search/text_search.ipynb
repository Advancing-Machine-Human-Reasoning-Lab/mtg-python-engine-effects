```python
import pandas as pd
```

```python
''' OPENING CSV '''

df = pd.read_csv('allcards.csv')
df = df.dropna(subset=['text']) # comment if you want cards with no text
df
```

```python
''' FILTERING DF '''

# replace the string in contains() with whatever you want to search for
# regex in contains() is true by default, so you can use regex
filtered = df[df['text'].str.contains('token')]#, case=False)] # uncomment if you don't want it case sensitive
for index, row in filtered.iterrows():
    print(row['name'], '---', row['text'])
    print('\n')
```
